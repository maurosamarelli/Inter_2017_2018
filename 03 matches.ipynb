```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.io.json import json_normalize
import util
path = "C:/Users/Mauro/OneDrive/Documenti/Football/playerank-master"
inter_id = 3161
```

### matches

**GET ALL SERIE A MATCHES**

```python
matches = pd.read_json(os.path.join(path, 'data/matches/matches_Italy.json'))
len(matches)
matches.head(3)
```

**GET ALL INTER MATCHES**

```python
inter_matches = matches[matches.label.str.contains("Internazionale")]
inter_matches.head(3)
```

**NORMALIZE REFEREES DATA**

```python
inter_matches = util.get_referees(inter_matches, "referees")
inter_matches.head(3)
```

**GET TEAMS DATA FOR EVERY INTER MATCH IN A DICTIONARY**

```python
teamsData = util.get_teamsData(inter_matches)
match_id = inter_matches[inter_matches.label.str.contains("Lazio - Internazionale")].wyId[0]
teamsData[match_id]
```

**GET ALL THE FORMATIONS ABOUT INTER MATCHES (OPPONENT TOO)**

```python
formationsData = util.get_formations(teamsData)
match_id = inter_matches[inter_matches.label.str.contains(" - Milan")].wyId.iloc[0]
formationsData[formationsData.matchId == match_id]
```

**EXPLORE INFORMATION ABOUT INTER MATCHES**
